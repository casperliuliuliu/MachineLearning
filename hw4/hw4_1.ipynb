```python
import torch
from torch.optim import Adam

def find_min_quartic(a4, a3, a2, a1, a0, initial_x=0.0, lr=0.1, tolerance=1e-10, max_iters=1000):
    # 定義方程式
    def equation(x):
        return a4 * x**4 + a3 * x**3 + a2 * x**2 + a1 * x + a0
    
    # # 定義梯度
    # def gradient(x):
    #     return 4 * a4 * x**3 + 3 * a3 * x**2 + 2 * a2 * x + a1
    
    # 初始化x值
    x = torch.tensor(initial_x, requires_grad=True)
    
    # 定義優化器
    optimizer = Adam([x], lr=lr)
    
    # 開始遞迴尋找最小值
    for iteration in range(max_iters):
        # 計算方程式的值和梯度
        y = equation(x)
        # grad = gradient(x)
        
        # 使用優化器更新x值
        optimizer.zero_grad()
        y.backward()
        optimizer.step()
        
        # # 檢查是否收斂
        # if torch.abs(grad) < tolerance:
        #     break
    
    # 返回最小值的x值和對應的方程式值
    return x.item(), equation(x).item()

# 測試函數
a4 = 1.0  # 注意: 這裡假設 a4 > 0
a3 = -4.0
a2 = 6.0
a1 = -4.0
a0 = 1.0

xmin, min_value = find_min_quartic(a4, a3, a2, a1, a0)
print(f"最小值 x_min: {xmin}, 方程式值: {min_value}")
```

```python
import torch

def find_min_recursive(a4, a3, a2, a1, a0, initial_x=0.0, lr=0.1, tolerance=1e-10, max_iters=1000):
    # 定義方程式
    def equation(x):
        return a4 * x**4 + a3 * x**3 + a2 * x**2 + a1 * x + a0
    
    # 定義梯度
    def gradient(x):
        return 4 * a4 * x**3 + 3 * a3 * x**2 + 2 * a2 * x + a1
    
    # 初始化x值
    x = torch.tensor(initial_x, requires_grad=True)
    
    # 開始遞迴尋找最小值
    for iteration in range(max_iters):
        # 計算方程式的值和梯度
        y = equation(x)
        grad = gradient(x)
        
        # 使用梯度下降更新x值
        x = x - lr * grad
        
        # 檢查是否收斂
        if torch.abs(grad) < tolerance:
            break
    
    # 返回最小值的x值和對應的方程式值
    return x.item(), equation(x).item()

# 測試函數
a4 = 1.0  # 注意: 這裡假設 a4 > 0
a3 = -4.0
a2 = 6.0
a1 = -4.0
a0 = 1.0

xmin, min_value = find_min_quartic(a4, a3, a2, a1, a0)
print(f"最小值 x_min: {xmin}, 方程式值: {min_value}")
```
