```python
import torch

def find_min_recursive(a4, a3, a2, a1, a0, initial_x=0.0, lr=0.1, tolerance=1e-10, max_iters=1000):
    # 
    def equation(x):
        return a4 * x**4 + a3 * x**3 + a2 * x**2 + a1 * x + a0
    
    # 
    def gradient(x):
        return 4 * a4 * x**3 + 3 * a3 * x**2 + 2 * a2 * x + a1
    
    # x
    x = torch.tensor(initial_x, requires_grad=True)
    
    # 
    for iteration in range(max_iters):
        # 
        y = equation(x)
        grad = gradient(x)
        
        # x
        x = x - lr * grad
        
        # 
        if torch.abs(grad) < tolerance:
            break
    
    # x
    return x.item(), equation(x).item()

# 
a4 = 1.0  # :  a4 > 0
a3 = -4.0
a2 = 6.0
a1 = -4.0
a0 = 1.0

xmin, min_value = find_min_quartic(a4, a3, a2, a1, a0)
print(f" x_min: {xmin}, : {min_value}")
```
